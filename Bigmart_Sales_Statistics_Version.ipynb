### Encoding & Feature Scaling

Is it Important to Split the Data back into train and test before Scaling?

Yes, this is to ensure that there is no data leakage. We cannot expose our test set with the training set when it comes to model building or scaling the data.

How to ensure this?

* We apply ``fit_transform`` in train when we scale the training data and we apply ``.transform`` when scaling the test set.

When fi_transform works, it will calculate the mean and standard deviation of each column and will do the transformation. So, Standard Scaler model has learnt from the Training Set and what it has learnt - **mean and sd**.

Now, the model will use the mean and standard deviation of the training set to transform the test set. This is the correct approach while scaling the data.

```python
# Divide the Data in X and y

# X represents input for model and y is the target/output for the model

X = newtrain.drop("Item_Outlet_Sales", axis = 1)
y = newtrain.Item_Outlet_Sales

X.columns, newtest.columns
```

```python
# Extract the Numerical Columns
num_cols = X.select_dtypes(include = np.number).columns
```

```python
# Scaling the train Set

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

for i in num_cols:
  X.loc[:, i] = scaler.fit_transform(pd.DataFrame(X.loc[:, i])) # Scaler learns mean and sd
  newtest.loc[:, i] = scaler.transform(pd.DataFrame(newtest.loc[:, i])) # .transform will use the mean and sd to transform test set
```

### About Standard Scaler

* Standard Scaler is known as Z Score Transformation. It takes the data substract it from mean and divide it by Standard Deviation.

* The values obtained are called Z Score Values. We scale the data in such a way such that the Mean is 0 and Standard Deviation is 1.

### Example of Standard Scaler

```python
# Example of Scaling

df = pd.DataFrame({"Col1": [1, 2, 3, 4, 10],
                   "Col2":[100, 250, 129, 234, 109]})
```

```python
# Manual Approach: (x-mew)/sd

mew_col1 = df.Col1.mean()
sd_col1 = df.Col1.std(ddof = 0) # Divide it by N as Python by default divides by n-1
df["trf_col1"] = (df.Col1 - mew_col1)/sd_col1
```

```python
# Col2:
mean_col2 = df.Col2.mean()
sd_col2 = df.Col2.std(ddof = 0) # Divide it by N as Python by default divides by n-1

df["trf"] = (df.Col2 - mean_col2)/sd_col2
```

```python
df.describe()
```

```python
sc = StandardScaler()

df["Scaled1"] = sc.fit_transform(pd.DataFrame(df.loc[:, 'Col1']))
df["Scaled2"] = sc.fit_transform(pd.DataFrame(df.loc[:, 'Col2']))
```

```python
df.head()
```

```python
# Min Max Scaler:: It is used in Image Processing

from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()

df["minmax1"] = minmax.fit_transform(pd.DataFrame(df.Col1))
df["minmax2"] = minmax.fit_transform(pd.DataFrame(df.Col2))

# Min Max Scales the Data in such a way that each col has min value 0 and Max as 1.
# It is never used when there is a presence of Outliers.
# Tip: Use Robust Scaler when there is a presence of Outlier in the Data
```

### Transformation

* Why Transform?

Most of the Machine Learning Models have an assumption that the Data should be Normally Distributed. For example, **Linear Regression** and **Logistic Regression** followed by KNNs are some of the model which expect the data to be Scaled and Normalized.

* This is the reason Transformation is applied on the data.

**Different Types of Transformations**

* Log Transformation/Exponential Transformation
* Log Normal Transformation (Finance)
* Square Root Transformation
* **Box Cox Transformation**
* **Yeo-Johnson Transformation**

The objective of the transformation is to make sure that the model is able to capture the pattern very well.

### How to Decide if we need to do any Scaling or Transformation

* Models like Linear Regression, Logistic Regression, KNN need Transformation
* Models like Linear Regression, Logistic Regression, KNN & Neural Net expect us to Scale the Data before Building Model

* Objective of Transformation is to reduce the skewness so if the data is not skewed or very light skewed, you dont need transformation and you can build the model directly.

* Models like Decision Tree, Random Forest, Gradient Boosting, Xtreme Gradient Boosting Machines dont expect you to either scale or Transform Data. You can directly take the data and build the model and do the predictions.

```python
nums = ['Item_Weight', 'Item_Visibility', 'Item_MRP',
       'Magic1', 'Magic2', 'Magic3', 'Magic4', 'Magic5', 'Magic6', 'Magic7',
       'Magic8', 'Magic9', 'Magic10', 'Magic11', 'Magic12']

# Num vs Num => T Test
# Ho: Predictors and Target Var are not Related/Independent.
# In Simple words, Item MRP has no relation with the Sales
# Ha: Predictors and target are related.

predictor = newtrain.loc[:, "Item_MRP"]

import scipy.stats as stats
stats.ttest_ind(predictor, newtrain.Item_Outlet_Sales)
```

### Anova

```python
# List of the Categorical Vars
newtrain.select_dtypes(include = "object").columns
```

```python
# Ho: Item Fat Content has no relation with the Target
# Ha: Ho is False. They are Related

import statsmodels.formula.api as sfa
from statsmodels.stats.anova import anova_lm

model = sfa.ols("Item_Outlet_Sales~Item_Fat_Content+Outlet_Size+\
Outlet_Location_Type+Outlet_Type+Item_ID+Item_Type_Cat+Outlets_ID",
                data = newtrain).fit()

print(anova_lm(model))
```

```python
newtrain.head()
```

```python
# Item Fat Content vs Item ID

# Ho: Item ID and Fat Content are not Related/Independent
# Ha: Ho is False

tbl = pd.crosstab(newtrain.Item_Fat_Content, newtrain.Item_ID)

# Chi Square Test
teststats, pvalue, df, expected_values = stats.chi2_contingency(tbl)
print(pvalue)
```

```python
# Outlet_Type vs Item ID

# Ho: Item ID and Outlet_Type are not Related/Independent
# Ha: Ho is False
tbl = pd.crosstab(newtrain.Outlet_Type, newtrain.Item_ID)

# Chi Square Test
teststats, pvalue, df, expected_values = stats.chi2_contingency(tbl)
print(pvalue)
```
